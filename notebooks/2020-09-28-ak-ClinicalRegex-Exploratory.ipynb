## Texts are not the same

```python
def list_texts_by_id(df):
    return df.groupby('id')\
             .agg({'text': lambda values: sorted(list(values)),
                   'annotationValues': list})\
             .reset_index()
```

```python
df_tk = df_k.pipe(list_texts_by_id)
df_ts = df_s.pipe(list_texts_by_id)
```

```python
(df_tk['id'] == df_ts['id']).value_counts()
```

```python
(df_tk['text'] == df_ts['text']).value_counts()
```

```python
df_t = df_tk.merge(df_ts,
                   left_on='id',
                   right_on='id',
                   suffixes=('_k', '_s'))

df_t['has_same_texts'] = df_t['text_k'] == df_t['text_s']
```

```python
# different texts
df_dt = df_t[~df_t['has_same_texts']]
```

```python
for index, example in df_dt.iterrows():
    print(index, example['id'], len(example['text_k']), len(example['text_s']))
```
