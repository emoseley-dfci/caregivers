```python
from tableone import TableOne

from cleaning.caregivers import (
    load_data,
    get_hadm_data
)
```

```python
df = load_data()
df_h = get_hadm_data(df)
```

```python
columns=[
    "ADMISSION_AGE",
    "GENDER",
    "MARITAL_STATUS",
    "DISCHARGE_LOCATION",
    "ETHNICITY",
    "LOS_HADM",
    "HOSPITAL_EXPIRE_FLAG",
    "CAR",
    "LIM"
]

nonnormal = [
    "LOS_HADM",
    "ADMISSION_AGE"
]

rename = {
    "GENDER": "SEX",
    "HOSPITAL_EXPIRE_FLAG": "SURVIVED_HADM"
}
```

# Table 1
Total study population characteristics

```python
TableOne(
    df_h,
    columns = columns,
    nonnormal = nonnormal,
    rename = rename
)
```

# Table 2
Study population stratified by child, spouse, and both (unadjusted)

```python
TableOne(
    df_h,
    columns = columns,
    groupby = "ANNOTATION_CHILD",
    nonnormal = nonnormal,
    rename = rename,
    pval = True
)
```

```python
TableOne(
    df_h,
    columns = columns,
    groupby = "ANNOTATION_SPOUSE",
    nonnormal = nonnormal,
    rename = rename,
    pval = True
)
```

```python
TableOne(
    df_h,
    columns = columns,
    groupby = "ANNOTATION_BOTH",
    nonnormal = nonnormal,
    rename = rename,
    pval = True
)
```
