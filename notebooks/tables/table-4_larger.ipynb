```python
import pandas as pd
from tableone import TableOne

from cleaning.caregivers.main import load_data
from notebooks.tables.ref import columns, nonnormal

df = load_data()

df_survived = df[~df["HOSPITAL_EXPIRE_FLAG"]]


annotation_child = TableOne(
    df_survived,
    columns = columns,
    groupby = "ANNOTATION_CHILD",
    nonnormal = nonnormal
)

annotation_spouse = TableOne(
    df_survived,
    columns = columns,
    groupby = "ANNOTATION_SPOUSE",
    nonnormal = nonnormal
)

annotation_both = TableOne(
    df_survived,
    columns = columns,
    groupby = "ANNOTATION_BOTH",
    nonnormal = nonnormal
)

annotation_any = TableOne(
    df_survived,
    columns = columns,
    groupby = "ANNOTATION_ANY",
    nonnormal = nonnormal
)


table = pd.concat([
    annotation_child.tableone,
    annotation_spouse.tableone,
    annotation_both.tableone,
    annotation_any.tableone
], axis=1)
```
