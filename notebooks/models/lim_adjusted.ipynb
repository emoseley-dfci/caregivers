# Models: LIM Documentation, Adjusted

```python
from cleaning.caregivers.models import (
    load_data,
    run_logit,
    format_logit_results
)

from notebooks.models.ref import FORMULA_CONTROL

df = load_data()
```

## Caregiver:Child[Binary]

```python
formula = "IDENTIFIED_CONV_LIM ~ {control_variables} + ANNOTATION_CHILD".format(control_variables=FORMULA_CONTROL)

model = run_logit(formula, df)

format_logit_results(model)
```

## Caregiver:Spouse[Binary]

```python
formula = "IDENTIFIED_CONV_LIM ~ {control_variables} + ANNOTATION_SPOUSE".format(control_variables=FORMULA_CONTROL)

model = run_logit(formula, df)

format_logit_results(model)
```

## Caregiver:Child[Binary] + Caregiver:Spouse[Binary]

```python
formula = "IDENTIFIED_CONV_LIM ~ {control_variables} + ANNOTATION_CHILD + ANNOTATION_SPOUSE".format(control_variables=FORMULA_CONTROL)

model = run_logit(formula, df)

format_logit_results(model)
```

## Caregiver[Categorical]

```python
formula = "IDENTIFIED_CONV_LIM ~ {control_variables} + C(ANNOTATION, Treatment(reference='NEITHER'))".format(control_variables=FORMULA_CONTROL)

model = run_logit(formula, df)

format_logit_results(model)
```
